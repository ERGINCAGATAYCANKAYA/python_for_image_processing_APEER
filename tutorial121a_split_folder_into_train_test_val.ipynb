```python
import splitfolders  
input_folder = '/content/drive/MyDrive/Colab Notebooks/data/sandstone/128_patches'
output_folder='/content/drive/MyDrive/Colab Notebooks/data/sandstone/128_patches_for_dataaug_video121'
```

```python


# Split with a ratio.
# To only split into training and validation set, set a tuple to `ratio`, i.e, `(.8, .2)`.
splitfolders.ratio(input_folder, output=output_folder, seed=42, ratio=(.8, .2), group_prefix=None) # default values

# Split val/test with a fixed number of items e.g. 100 for each set.
# To only split into training and validation set, use a single number to `fixed`, i.e., `10`.
#splitfolders.fixed("input_folder", output="data2", seed=1337, fixed=(100, 100), oversample=False, group_prefix=None) # default values
```
